```python
import numpy as np

def single_bandit(n_arm, q_true, q_est, epsilon, steps, alpha=None, stationary=True, ucb_c=None):
    act_counts = np.zeros(n_arm)
    rewards = list()
    opt_act = np.argmax(q_true)
    opt_acts = list()
    for i in range(steps):
        # for exercise 2.5
        if not stationary:
            q_true += np.random.randn(n_arm) * 0.01
            opt_act = np.argmax(q_true)
        # for figure 2.4
        if ucb_c:
            a = np.argmax(q_est + ucb_c * np.sqrt(np.log(i + 1) / (act_counts + 1e-5)))
        elif np.random.rand() < epsilon:
        	a = np.random.choice(n_arm)
        else:
        	a = np.argmax(q_est)
        act_counts[a] += 1
        reward = q_true[a] + np.random.randn()
        rewards.append(reward)
        opt_acts.append(int(a == opt_act))
        # for exercise 2.5
        if alpha:
            q_est[a] += alpha * (reward - q_est[a])
        else:
            q_est[a] += 1 / act_counts[a] * (reward - q_est[a])
    return rewards, opt_acts
```

```python
import matplotlib.pyplot as plt

n_arm = 10
rewards = {}
opt_acts = {}
for epsilon in [0.0, 0.01, 0.1]:
	res = [single_bandit(n_arm, np.random.randn(n_arm), np.zeros(n_arm), epsilon, 1000, alpha=None, stationary=True) for i in range(2000)]
	rs_, as_ = np.array(list(zip(*res))).mean(axis=1)
	rewards[epsilon] = rs_
	opt_acts[epsilon] = as_


plt.figure(figsize=(10, 12))

plt.subplot(2, 1, 1)
for k, v in rewards.items():
	plt.plot(v, label=f'epsilon = {k}')
plt.xlabel('steps')
plt.ylabel('average reward')
plt.legend()

plt.subplot(2, 1, 2)
for k, v in opt_acts.items():
	plt.plot(v, label=f'epsilon = {k}')
plt.xlabel('steps')
plt.ylabel('% optimal action')
plt.legend()
plt.show()
```

```python
def max_w_random(n_arm, c=1):
	"""
	This shows why the spike's size is about 0.43. 
	Basically to check how often you get the true max when you add a random number to it.
	"""
	q_true = np.random.randn(n_arm)
	oa_true = np.argmax(q_true)
	oa_w_random = np.argmax(q_true + c * np.random.randn(n_arm))
	return oa_true == oa_w_random

n_sample = 20000
ts = [max_w_random(10, 1) for i in range(n_sample)]
print(np.mean(ts))
```

```python
import math
def get_limit(n_arm, n_sample=10000):
	"""
	This show why the limit is around 1.54.
	Check the mean of the max of n_arm random variables.
	Theoretically: 
	1. https://stats.stackexchange.com/questions/381212/distribution-of-maximum-of-normally-distributed-random-variables
	2. https://math.stackexchange.com/questions/275414/distribution-of-largest-sample-from-normal-distribution (some close forms)
	"""
	x = np.random.randn(n_sample, n_arm)
	return np.max(x, axis=1).mean()

print(get_limit(3, 10000))
print(3 / 2 / math.sqrt(math.pi))
```

```python
import numpy as np

def softmax(x):
    return(np.exp(x)/np.exp(x).sum())

def single_bandit(n_arm, q_true, q_est, steps, epsilon=None, alpha=None, stationary=True, ucb_c=None, gradient=None, gradient_baseline=None):
    act_counts = np.zeros(n_arm)
    rewards = list()
    opt_act = np.argmax(q_true)
    opt_acts = list()
    hs = np.zeros(n_arm)
    for i in range(steps):
        # 1. SELECT ACTION
        # for exercise 2.5
        if not stationary:
            q_true += np.random.randn(n_arm) * 0.01
            opt_act = np.argmax(q_true)
        if gradient:
            ps = softmax(hs)
            a = np.random.choice(n_arm, p=ps)
        # for figure 2.4
        elif ucb_c:
            a = np.argmax(q_est + ucb_c * np.sqrt(np.log(i + 1) / (act_counts + 1e-5)))
        elif np.random.rand() < epsilon:
        	a = np.random.choice(n_arm)
        else:
        	a = np.argmax(q_est)
        act_counts[a] += 1
        reward = q_true[a] + np.random.randn()
        rewards.append(reward)
        opt_acts.append(int(a == opt_act))
        # 2. UPDATE Q_EST or Preference (for gradient)
        # figure 2.5
        if gradient:
            one_hot = np.zeros(n_arm)
            one_hot[a] = 1
            baseline = np.mean(rewards) if gradient_baseline else 0
            hs += alpha * (reward - baseline) * (one_hot - ps) 
        else:
            if alpha:
            # for exercise 2.5
                q_est[a] += alpha * (reward - q_est[a])
            else:
                q_est[a] += 1 / act_counts[a] * (reward - q_est[a])
    return rewards, opt_acts
```

```python
import matplotlib.pyplot as plt
from itertools import product

n_arm = 10
plt.figure(figsize=(10, 6))
for baseline, alpha in product([None, True], [0.1, 0.4]):
	res = [single_bandit(n_arm, np.random.randn(n_arm) + 4, None, 1000, None, alpha, True, None, True, baseline) for i in range(2000)]
	rs_, as_ = np.array(list(zip(*res))).mean(axis=1)
	plt.plot(as_, label=f'baseline = {baseline}, alpha = {alpha}')
plt.xlabel('steps')
# plt.ylabel('average reward')
plt.ylabel('% optimal action')
# plt.xlim(-10, 40)
plt.legend()
plt.show()
```
