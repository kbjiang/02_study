### Two pointers
1. Pattern: *move right pointer till condition met, then move left pointer till condition break*
1. use `while` to move left pointer is much cleaner than using indices
1. use `.add/.remove` to update temp iterables instead of slicing; see comment in example.
1. Use start and end indices instead of slicing--slicing and concatenating arrays would take $O(n)$ time.

```python
# 209, Minimum size subarray sum
def minSubArrayLen(target: int, nums):
    r = 0
    curList = []

    minLen = len(nums)
    # Had I used slicing, the right boundary condition would be weird.
    # for r in range(len(nums)+1):
    #     curList= nums[l:r]
    for r in range(len(nums)): 
        curList.append(nums[r])
        while sum(curList) >= target:
            minLen = min(minLen, len(curList))
            curList = curList[1:]
            # print(curList)
    if minLen == len(nums) and sum(nums) < target:
        return 0
    return minLen
```

```python
# 76. Minimum Window Substring
def minWindow(s, t):
    l = 0
    r = 0
    minWin = "x"*(max(len(s), len(t))+1)
    t_counts = {char: t.count(char) for char in set(t)}

    for r in range(len(s)):
        if s[r] in t_counts:
            t_counts[s[r]] -= 1
        while all([cnt <= 0 for cnt in t_counts.values()]):
            minWin_ = s[l:r+1]
            minWin = min(minWin, minWin_, key=len)
            if s[l] in t_counts:
                t_counts[s[l]] += 1
            l += 1
    if minWin == "x"*(max(len(s), len(t))+1):
        return ""
    return minWin
```

```python
# 121. Best time to buy and sell stock
def maxProfit(prices:int) -> int:
    if len(prices) <= 1:
        return 0
    l = 0
    max_profit = 0
    for r in range(1, len(prices)):
        profit = prices[r] - prices[l]
        if profit < 0:
            l = r
            continue
        elif profit > max_profit:
            print(l, r)
            max_profit = profit
    # boundary condition
    if l == len(prices) - 1:
        return 0
    return max_profit
```

```python
# 238 product of array except self
def productExceptSelf(nums):
    def get_products(nums_):
        product_ = 1
        products = []
        for num in nums_:
            products.append(product_)
            product_ = product_ * num
        return products
    products_prefix = get_products(nums)
    products_suffix = get_products(nums[::-1])[::-1]

    return [x * y for x, y in zip(products_prefix, products_suffix)]
```

### hash table

**Use *Two Sum* as an example to show procedure**
1. "exactly one solution" means can stop early and no worries about running outta boundary
1. no "same element twice" but different elements can have same value
1. edge case:
    1. diff elem with same value can lead to collison, so we should search before insert

```python
# 1. Two sum
def twoSum(nums, target):
    tbl = dict()
    for i, num in enumerate(nums):
        candidate = target - num
        if candidate in tbl:  # O(1)
            return tbl[candidate], i
        tbl[num] = i
```

```python
nums = [2,7,11,15]
target = 9
nums = [3, 2, 4]
target = 6
nums = [3, 3]
target = 6
twoSum(nums, target)
```
